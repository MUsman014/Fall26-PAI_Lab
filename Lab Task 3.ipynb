```python


jug1_capacity = 4
jug2_capacity = 3
goal = 2

visited = set()

def dfs(jug1, jug2, path):

    if (jug1, jug2) in visited:
        return False

    visited.add((jug1, jug2))
    path.append((jug1, jug2))


    if jug1 == goal or jug2 == goal:
        print("Solution Found!\nSteps:")
        for step in path:
            print(step)
        return True

    
    moves = [
        (jug1_capacity, jug2, "Fill Jug1"),
        (jug1, jug2_capacity, "Fill Jug2"),
        (0, jug2, "Empty Jug1"),
        (jug1, 0, "Empty Jug2"),
        (jug1 - min(jug1, jug2_capacity - jug2), jug2 + min(jug1, jug2_capacity - jug2), "Pour Jug1 -> Jug2"),
        (jug1 + min(jug2, jug1_capacity - jug1), jug2 - min(jug2, jug1_capacity - jug1), "Pour Jug2 -> Jug1")
    ]


    for new_jug1, new_jug2, rule in moves:
        if (new_jug1, new_jug2) not in visited:
            
            print(f"Rule Applied: {rule} -> ({new_jug1}, {new_jug2})")
            if dfs(new_jug1, new_jug2, path.copy()):
                return True

    return False


dfs(0, 0, [])
```
