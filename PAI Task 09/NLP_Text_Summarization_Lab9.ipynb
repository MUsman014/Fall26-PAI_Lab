## Text Summarization (Python)

```python
from collections import Counter
import re

text = """Natural Language Processing is a field of Artificial Intelligence.
It helps computers understand human language.
It is used in chatbots, translation, and sentiment analysis.
NLP is very useful in modern applications."""

sentences = re.split(r'[.!?]', text)

words = text.lower().split()

word_freq = Counter(words)

sentence_scores = {}

for sentence in sentences:
    for word in sentence.lower().split():
        if word in word_freq:
            if sentence not in sentence_scores:
                sentence_scores[sentence] = word_freq[word]
            else:
                sentence_scores[sentence] += word_freq[word]

summary = max(sentence_scores, key=sentence_scores.get)

print("Original Text:\n", text)
print("\nSummary:\n", summary.strip())
```
