```python
def is_position_safe(chessboard, current_row, current_col, size):
    
    for r in range(current_row):
        if chessboard[r][current_col] == 1:
            return False

    
    r, c = current_row - 1, current_col - 1
    while r >= 0 and c >= 0:
        if chessboard[r][c] == 1:
            return False
        r -= 1
        c -= 1

    
    r, c = current_row - 1, current_col + 1
    while r >= 0 and c < size:
        if chessboard[r][c] == 1:
            return False
        r -= 1
        c += 1

    return True

def solve_n_queens_recursive(chessboard, current_row, size, all_solutions):
    if current_row == size:
        
        formatted_solution = [
            "".join("Q" if cell == 1 else "." for cell in row)
            for row in chessboard
        ]
        all_solutions.append(formatted_solution)
        return

    for col_index in range(size):
        if is_position_safe(chessboard, current_row, col_index, size):
            chessboard[current_row][col_index] = 1
            solve_n_queens_recursive(chessboard, current_row + 1, size, all_solutions)
            chessboard[current_row][col_index] = 0 

def solve_n_queens_board(size):
    chessboard = [[0 for _ in range(size)] for _ in range(size)]
    all_solutions = []
    solve_n_queens_recursive(chessboard, 0, size, all_solutions)
    return all_solutions


board_size = 8
solutions = solve_n_queens_board(board_size)
print(f"Total solutions for {board_size}-Queens: {len(solutions)}\n")
for solution in solutions:
    for row in solution:
        print(row)
    print()
```
